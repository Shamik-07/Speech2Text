# Making video clips from YouTube and extracting the audio from them

```python
from IPython.display import Audio, YouTubeVideo, display
import moviepy.editor as mp
import os
```

```python
display(YouTubeVideo(id="zE7PKRjrid4"))
```

```python
# Substitute below youtube link
clip = "https://www.youtube.com/watch?v=zE7PKRjrid4"

# Substitue below for start/end seconds
start = 0
end = 127
```

```python
# Download the clip as mp4 & rename it; this will download the clip in the current directory and rename it to "clip.mp4"
os.system('youtube-dl {} --recode-video mp4'.format(clip))
os.system('mv *.mp4 clip.mp4')
```

## Making a smaller clip from the YouTube video

```python
clip = mp.VideoFileClip("clip.mp4").subclip(0,127)
```

```python
clip.write_videofile("clip_short.mp4")
```

## Extract Audio

**Saving the audio in multiple 10 sec audio files**

```python
clip = mp.VideoFileClip("clip.mp4")
end = min(clip.duration, end)

# Save the paths for later
clip_paths = []

# Extract Audio-only from mp4
for i in range(start, int(end), 10):
  sub_end = min(i+10, end)
  sub_clip = clip.subclip(i,sub_end)

  sub_clip.audio.write_audiofile("audio_" + str(i) + ".mp3")
  clip_paths.append("audio_" + str(i) + ".mp3")
```

**Extracting and saving the audio file of the shorter video**

```python
clip = mp.VideoFileClip("clip.mp4").subclip(0,127)
clip.audio.write_audiofile("audio.mp3")
```

```python
# Playing the saved Audio from the current directory
Audio("audio.mp3")
```
